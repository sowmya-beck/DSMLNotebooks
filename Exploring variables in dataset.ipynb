```python
import seaborn as sns
import pandas as pd

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# Selecting categorical features
categorical_features = titanic.select_dtypes(include=['object', 'category']).columns.tolist()
print("Categorical Features:", categorical_features)
```

```python
# Selecting discrete features (integer types)
discrete_features = titanic.select_dtypes(include=['int64']).columns.tolist()
print("Discrete Features:", discrete_features)
```

```python
# Selecting continuous features (float types)
continuous_features = titanic.select_dtypes(include=['float64']).columns.tolist()
print("Continuous Features:", continuous_features)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting categorical feature: 'sex'
sns.countplot(x='sex', data=titanic)
plt.title('Distribution of Passengers by Sex')
plt.show()
```

```python
# Plotting discrete feature: 'sibsp' (siblings/spouses aboard)
sns.countplot(x='sibsp', data=titanic)
plt.title('Number of Siblings/Spouses Aboard')
plt.show()
```

```python
# Plotting continuous feature: 'age'
sns.histplot(titanic['age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Passenger Ages')
plt.show()
```
